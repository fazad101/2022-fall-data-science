* Plot a histogram of the base_salary.  Write one sentence about what this tells you about base salaries.

* Make a boxplot of base salaries per (or by) work_location_borough, what does this tell you about the difference in working in each location?
* Make a plot visualizing the difference average base salaries based on agency_name.  Who would you want to work for based on this information?

```python
df = pd.read_csv('data/budget-sample.csv')
df.head(10)
```

```python
# Plot a histogram of the base_salary.
df['base_salary'].plot(kind='hist')
```

```python
# Write one sentence about what this tells you about base salaries
print("Most people's base salary is around 80,000")
```

```python
# Make a boxplot of base salaries
df['base_salary'].plot(kind='box')
```

```python
# what does this tell you about base salaries?
print('There are many outliers above the maximum.')
```

```python
# Make a boxplot of base salaries per (or by) work_location_borough
ax = sns.boxplot(data=df, x='base_salary', y='work_location_borough')
```

```python
# What does this tell you about the difference in working in each location?
print('Manhattan is the best.')
```

```python
# Make a plot visualizing the difference average base salaries based on agency_name.
gb = df.groupby('agency_name')
gb['base_salary'].mean().sort_values().plot(kind='bar', figsize=(15, 10))
```

```python
# Who would you want to work for based on this information?
print('NYC Fire Pension Fund')
```

```python
# Tell me something new, cool, and interesting about this data
sns.pairplot(df)
```

# Using the BX-Police.csv data:
* create a time series chart illustrating their average base salary over time (by fiscal_year).  
* What trend does this show? 
* Why do you think the trend exists?


```python
df = pd.read_csv('data/BX-Police.csv')
df.head()
```

```python
# create a time series chart illustrating their average base salary over time (by fiscal_year).
gb = df.groupby('fiscal_year')
gb['base_salary'].mean().plot(kind='line')
```

```python
# What trend does this show?
print('Salaries generally increase over time')
```

```python
# Why do you think the trend exists?
print('Inflation')
```

# Using Sales.csv 
* make a time series chart that shows the sales for each item over the past 12 months.  
* What does this tell you about the items.  
* Which item would you recommend they push more of and which item/s would you recommend they stop selling?
* show what percentage of each item contributes to total sales. 


```python
df = pd.read_csv('data/Sales.csv')
df.head()
```

```python
# make a time series chart that shows the sales for each item over the past 12 months.
features = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
df.plot(kind='line', x='month_number', y=features)
```

```python
# What does this tell you about the items.
print('Bathingsoup and toothpaste are the two most noticeable increase.')
```

```python
# Which item would you recommend they push more of and which item/s would you recommend they stop selling?
print('Only sell bathingsoup and toothpaste')
```

```python
# show what percentage of each item contributes to total sales.
products = df[features].sum()
products
```

```python
percentages = products/df[features].sum().sum()*100
percentages.sort_values().plot(kind='bar')
```
